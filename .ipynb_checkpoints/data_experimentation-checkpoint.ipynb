```python
train['region_en'] = train['region'].apply(lambda x : region_map[x])
train['parent_category_name_en'] = train['parent_category_name'].apply(lambda x : parent_category_name_map[x])
train['category_name_en'] = train['category_name'].apply(lambda x : category_map[x])

# bins of deal probability
interval = (-0.99, .10, .20, .30, .40, .50, .60, .70, .80, .90, 1.1)
cats = ['0-0.1', '0.1-0.2', '0.2-0.3', '0.3-0.4', '0.4-0.5', '0.5-0.6', '0.6-0.7', '0.7-0.8', '0.8-0.9','0.9-1.0']
train['deal_class'] = train['deal_probability'].apply(lambda x: ">=0.5" if x >=0.5 else "<0.5")
train["deal_class_2"] = pd.cut(train.deal_probability, interval, labels=cats)

# date time features
train['weekday'] = train.activation_date.dt.weekday
train['month'] = train.activation_date.dt.month
train['day'] = train.activation_date.dt.day
train['week'] = train.activation_date.dt.week
# english mapped of weekday
daymap = {0:'Sun', 1:'Mon', 2:'Tue', 3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat'}
train['weekday_en'] = train['weekday'].apply(lambda x : daymap[x])
```
